```python
import json
import csv
import os
```

```python
files = os.listdir('./images-data')

with open('data.csv', 'w') as csvfile:
    fieldnames = ['image_file', 'rating']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    with open('data.json', 'rb') as json_file:
        data_dict = json.loads(json_file.read())
        count = 1
        for row in data_dict:
            if (str(count) + '.png') in files:
                if row['rating'] is None:
                    rating = 0
                    developer = 0
                    developer_count = 0
                    creativity = 0
                    creativity_count = 0
                    design = 0
                    design_count = 0
                    usability = 0
                    usability_count = 0
                    content = 0
                    content_count = 0
                    mobile = 0
                    mobile_count = 0
                    for vote in row['votes']:
                        if 'developer_score' in vote.keys() and vote['developer_score'] is not None:
                            developer += float(vote['developer_score'])
                            developer_count += 1
                        if 'creativity_score' in vote.keys() and vote['creativity_score'] is not None:
                            creativity += float(vote['creativity_score'])
                            creativity_count += 1
                        if 'design_score' in vote.keys() and vote['design_score'] is not None:
                            design += float(vote['design_score'])
                            design_count += 1
                        if 'usability_score' in vote.keys() and vote['usability_score'] is not None:
                            usability += float(vote['usability_score'])
                            usability_count += 1
                        if 'content_score' in vote.keys() and vote['content_score'] is not None:
                            content += float(vote['content_score'])
                            content_count += 1
                        if 'mobile_score' in vote.keys() and vote['mobile_score'] is not None:
                            mobile += float(vote['mobile_score'])
                            mobile_count += 1
                    count_x = 0
                    developer_avg = 0
                    creativity_avg = 0
                    design_avg = 0
                    usability_avg = 0
                    content_avg = 0
                    mobile_avg = 0
                    rating = 0
                    if developer_count > 0:
                        developer_avg = developer / developer_count
                        rating += developer_avg
                        count_x += 1
                    if creativity_count > 0:
                        creativity_avg = creativity / creativity_count
                        rating += creativity_avg
                        count_x += 1
                    if design_count > 0:
                        design_avg = design / design_count
                        rating += design_avg
                        count_x += 1
                    if usability_count > 0:
                        usability_avg = usability / usability_count
                        rating += usability_avg
                        count_x += 1
                    if content_count > 0:
                        content_avg = content / content_count
                        rating += content_avg
                        count_x += 1
                    if mobile_count > 0:
                        mobile_avg = mobile / mobile_count
                        rating += mobile_avg
                        count_x += 1
                    rating = rating / count_x
                    writer.writerow({ 'image_file': str(count) + '.png', 'rating': rating })
                else:
                    writer.writerow({ 'image_file': str(count) + '.png', 'rating': row['rating']})
            count += 1
```
